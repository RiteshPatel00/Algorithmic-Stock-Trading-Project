### **Logarithmic Returns vs Arithmetic Returns**

Say you hold a stock as it increases from **\\$100** to **\\$105**. Usually, this is reported as a return of **5%**. This simple definition of return serves us well for most uses, but there are some quirks that make arithmetic returns difficult 
to use in some academic and professional evaluation settings. For example, continuously compounded arithmetic returns are not symmetric. 
If a position appreciates **15%** and then depreciates **15%**, the total change is **-2.25%**. To avoid this quirk, we need to sometimes use **log returns**

```python
log_returns = np.log(stocks/stocks.shift(1))
```

```python
log_returns.hist(bins = 100, figsize = (12,8))
plt.tight_layout()
```

```python
#Return in terms of logarithmic scale average
log_returns.mean()
```

```python
#Calculate the covariance which then is multiplied by 252 business days in a year 
log_returns.cov() * 252
```

**Want to add in random numbers in order to find expected returnn, volatility and sharpe ratios**

```python
np.random.seed(101)

print(stocks.columns)

weights = np.array(np.random.random(4))

print("Random Weights")
print(weights)

print('Rebalance')
#Normalize the weights to equal 1 when summed
weights = weights/np.sum(weights)

print(weights)

#Expected Return
print('Expected Portfolio Return')
exp_returns = np.sum((log_returns.mean() * weights) * 252)
print(exp_returns)

#Expected Volatility
print('Expected Volatility')
exp_volatility = np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 252, weights)))
print(exp_volatility)

#Sharpe Ratio
print('Sharpe Ratio')
SR = exp_returns / exp_volatility
print(SR)
```

**Now we basically want to run this 5000x in order to optimize our portfolio by doing a Monte Carlo Simulation**

```python
np.random.seed(101)

numberOfPortfolios = 5000
all_weights = np.zeros((numberOfPortfolios , len(stocks.columns)))
return_arr = np.zeros(numberOfPortfolios)
volatility_arr = np.zeros(numberOfPortfolios)
sharpeRatio_arr = np.zeros(numberOfPortfolios)


for index in range(numberOfPortfolios):

    #Weights
    weights = np.array(np.random.random(4))
    weights = weights/np.sum(weights)
    
    #Save Weights
    all_weights[index,:] = weights

    #Expected Return
    return_arr[index] = np.sum((log_returns.mean() * weights) * 252)

    #Expected Volatility
    volatility_arr[index] = np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 252, weights)))

    #Sharpe Ratio
    sharpeRatio_arr[index] = return_arr[index] / volatility_arr[index]
```

```python
#Find the max sharpe ratio
sharpeRatio_arr.max()
```

**The max Point in this case would be our ideal sharpe ratio, however this simulation happens to take up too much computing power and is quite slow, and for a larger number of portfolios it would not be ideal, which is why a optimization function can
be written**

```python
plt.figure(figsize = (12,8));
plt.scatter(volatility_arr,return_arr,c = sharpeRatio_arr, cmap='viridis');
plt.colorbar(label = 'Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')

#Label the max point
plt.scatter(volatility_arr[1420] , return_arr[1420], c = 'red', s= 50, edgecolors = 'black')
```

```python
def get_optimal_SR(weights):
    weights = np.array(weights)
    retrn = np.sum(log_returns.mean() * weights) * 252
    volatility = np.sqrt(np.dot(weights.T, np.dot(log_returns.cov() * 252, weights)))
    SR = retrn / volatility
    return np.array([retrn,volatility,SR])
```

```python
#Will help us using it's minimization function
from scipy.optimize import minimize
```

```python
#Helper functions
def negative_SR(weights):
    return get_optimal_SR(weights)[2] * -1

def check_sum(weights):
    #Returns zero if sum of weights is 1
    return np.sum(weights) - 1


constraints = ({'type' : 'eq' , 'fun' : check_sum})

#Values must be between 0 and 1
bounds = ((0,1),(0,1),(0,1),(0,1))

inital_guess = [0.25, 0.25, 0.25, 0.25]
```

```python
#SLSQP - Sequential Least Squares Programming Minimization technique
optimal_results = minimize(negative_SR , inital_guess, method = 'SLSQP', bounds = bounds, constraints=constraints)
```

```python
#Optimal Allocations
optimal_results.x
```

**As seen previously in the Monte Carlo Simulation, our optimal SR is in fact 1.03 but it's calculated with less computing power**

```python
#Returns [Daily Returns, Volatility, Optimal Sharpe Ratio]
get_optimal_SR(optimal_results.x)
```

**Now after getting the optimal results we can use Efficient Frontier which is the set of optimal portfolios 
that offer the highest expected return for a defined level of risk or the lowest risk for a given level of expected return**

```python
frontier_y = np.linspace(0,0.3,100)
```

```python
def minimize_volatility(weights):
    return get_optimal_SR(weights)[1]

frontier_volatility = []

for return_value in frontier_y:
    constraints = ({'type' : 'eq' , 'fun' : check_sum}, 
                   {'type' :  'eq' , 'fun' : lambda x : get_optimal_SR(x)[0] - return_value})
    
    result = minimize(minimize_volatility, inital_guess, method='SLSQP',bounds=bounds, constraints=constraints)
    
    frontier_volatility.append(result['fun'])
    
```

### Anything that lies below the Efficient Fronteir is considered sub-optimal for a given return or volatility

```python
#Scatter plot of SR Values
plt.figure(figsize = (12,8));
plt.scatter(volatility_arr,return_arr,c = sharpeRatio_arr, cmap='viridis');
plt.colorbar(label = 'Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')

#Optimal SR Values for Portfolios
plt.plot(frontier_volatility,frontier_y,'b--')
```
