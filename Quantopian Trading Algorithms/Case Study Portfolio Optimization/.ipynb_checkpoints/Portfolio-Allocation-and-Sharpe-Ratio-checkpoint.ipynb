```python
start = pd.to_datetime('2012-01-01')
end = pd.to_datetime('2017-01-01')
```

## Extract datasets from Quandl for Apple, Cisco, IBM and Amazon

```python
aapl = quandl.get('WIKI/AAPL.11',start_date=start,end_date=end)
csco = quandl.get('WIKI/CSCO.11',start_date=start,end_date=end)
ibm = quandl.get('WIKI/IBM.11',start_date=start,end_date=end)
amzn = quandl.get('WIKI/AMZN.11',start_date=start,end_date=end)
```

```python
aapl.iloc[0]['Adj. Close']
```

```python
for stock_df in (aapl,csco,ibm,amzn):
    stock_df['Normed Return'] = stock_df['Adj. Close'] / stock_df.iloc[0]['Adj. Close']
```

```python
amzn.tail()
```

### Lets allocate our money into these portfolios
* 30% into Apple
* 20% into Cisco
* 40% into Amazon
* 10% into IBM

```python
for stock_df, allo in zip((aapl,csco,ibm,amzn), [.3,.2,.4,.1]):
    stock_df['Allocation'] = stock_df['Normed Return']*allo
```

```python
aapl.head()
```

```python
#Allocate 1 million dollars to the portfolio

for stock_df in (aapl,csco,ibm,amzn):
    stock_df['Position Values'] = stock_df['Allocation']*1000000
```

```python
all_vals = [aapl['Position Values'],csco['Position Values'],ibm['Position Values'],amzn['Position Values']]

portfolio_val = pd.concat(all_vals,axis = 1)
```

```python
portfolio_val.columns = ['AAPL Position','CSCO Position','IBM Position','AMZN Position']
```

```python
#Get total position
portfolio_val['Total Position'] = portfolio_val.sum(axis = 1)
```

**From looking at the individual stock positions, it is clear to see that AMZN and APPL were good long term investments
while IBM and CSCO may have not been the best choice
We can use the sharpe ratio to figure out how much we should have allocated to each security**

```python
#Plot total value of position over time
portfolio_val['Total Position'].plot(figsize = (12,8));
plt.title("Total Portfolio Value")

#Plot positions
portfolio_val[['AAPL Position','CSCO Position','IBM Position','AMZN Position']].plot(figsize = (12,8));
```

```python
portfolio_val['Daily Return'] = portfolio_val['Total Position'].pct_change(1)
```

```python
portfolio_val['Daily Return'].mean()
```

```python
portfolio_val['Daily Return'].std()
```

```python
portfolio_val['Daily Return'].plot(kind = 'hist', bins = 100, figsize = (12,8),histtype='bar', ec='black');
portfolio_val['Daily Return'].plot(kind = 'kde',figsize = (12,8));
```

```python
cumulative_return = 100 * (portfolio_val['Total Position'][-1] / portfolio_val['Total Position'][0] - 1)
```

```python
#Cumulative return is about %84.74 
cumulative_return
```

**Now for the sharpe ratio assuming the risk free rate is equal to zero. 
Note that in the future the federal government can and will raise or drop interest rates**

```python
SR = portfolio_val['Daily Return'].mean() / portfolio_val['Daily Return'].std()
```

```python
SR
```

```python
#Square root by 252 since there are 252 business days for each of these stocks
annual_SR = (252 ** 0.5) * SR
```
