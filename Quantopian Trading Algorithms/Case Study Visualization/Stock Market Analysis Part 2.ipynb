```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_datareader
import datetime
import pandas_datareader as web
```

## Reading in Tesla,Ford and General Motors stock from datasets
---

```python
tesla = pd.read_csv("Datasets/Tesla_Stock.csv",index_col='Date',parse_dates=True)
ford = pd.read_csv("Datasets/Ford_Stock.csv",index_col='Date',parse_dates=True)
gm = pd.read_csv("Datasets/GM_Stock.csv",index_col='Date',parse_dates=True)
```

# Basic Financial Analysis
---
### Daily Percent Change
$ r_t = \frac{p_t}{p_{t-1}} -1$

```python
tesla['returns'] = tesla['Close'].pct_change(1)
ford['returns'] = ford['Close'].pct_change(1)
gm['returns'] = gm['Close'].pct_change(1)
```

**Plot histograms as a stack** \
Here it's clear to see that GM and Ford tend to be more stable and less volatile than Tesla

```python
tesla['returns'].hist(bins=100,label = "Tesla",figsize = (10,8), alpha = 0.5)
ford['returns'].hist(bins=100,label = "Ford",figsize = (10,8), alpha = 0.5)
gm['returns'].hist(bins=100,label = "GM",figsize = (10,8), alpha = 0.5)

plt.legend();
```

**We can also analyze through Kernel Density Estimation plots as well** \
From this we can derive that ford happens to be the most stable stock out of the three

```python
tesla['returns'].plot(kind='kde',label = "Tesla",figsize = (10,8))
ford['returns'].plot(kind='kde',label = "Ford",figsize = (10,8))
gm['returns'].plot(kind='kde',label = "GM",figsize = (10,8))

plt.legend();
```

**We can analyze further with the use of box plots to compare returns**

```python
box_df = pd.concat([tesla['returns'],ford['returns'],gm['returns']],axis = 1)
box_df.columns = ["Tesla Returns", "Ford returns", "GM returns"]
box_df.plot(kind = 'box',figsize = (10,12))
```

**Now lets compare daily returns between stocks through the use of a matrix scatter plot on the daily returns** \
Once again, we can see that there is some correlation on daily returns with GM and Ford 

```python
from pandas.plotting import scatter_matrix
scatter_matrix(box_df,figsize = (8,8), alpha = 0.2, hist_kwds = {'bins':100});
```

Since they do have some correlation, I'll plot both of them against each other in another cell
This seems fairly correlated and seems like a linear regression can be applied to further research the stocks

```python
box_df.plot(kind = 'scatter', x = 'Ford returns', y = 'GM returns' , alpha = 0.5, figsize = (8,8))
```

# Cumlative Returns 
---
**Useful to give investors a more immediate insigt into the 
gains made by the stock from the inital purchase date so let's plot that out as well** 

$ i_i = (1+r_t) * i_{t-1} $




```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
```

**Looking at the plot, it's clear to see that Tesla's Cumulative Return is greater, if you had bought stock from 2012
but this doesn't truly give the picture of Ford and GM so lets zoom in on their cumulative returns**

```python
tesla['Cumulative Return'].plot(label = 'Tesla', figsize = (16,8))
ford['Cumulative Return'].plot(label = 'Ford', figsize = (16,8))
gm['Cumulative Return'].plot(label = 'GM', figsize = (16,8))
```

**The cumulative returns also indicate correlation between Ford and GM as it was predicted with the correlation matrix**

```python
ford['Cumulative Return'].plot(label = 'Ford', figsize = (16,8))
gm['Cumulative Return'].plot(label = 'GM', figsize = (16,8))
```
