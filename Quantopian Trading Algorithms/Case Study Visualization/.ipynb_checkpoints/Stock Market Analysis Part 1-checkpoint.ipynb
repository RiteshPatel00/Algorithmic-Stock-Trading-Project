## Reading in Tesla,Ford and General Motors stock from datasets
---

```python
tesla = pd.read_csv("Datasets/Tesla_Stock.csv",index_col='Date',parse_dates=True)
ford = pd.read_csv("Datasets/Ford_Stock.csv",index_col='Date',parse_dates=True)
gm = pd.read_csv("Datasets/GM_Stock.csv",index_col='Date',parse_dates=True)
```

## Visualizing the Opening price of each stock
---

```python
tesla['Open'].plot(label="Tesla",figsize = (12,8), title = 'Opening Prices')
ford['Open'].plot(label="Ford")
gm['Open'].plot(label="General Motors")

plt.legend();
```

## Volume of Stock traded each day
---

```python
tesla['Volume'].plot(label="Tesla",figsize = (12,8), title = 'Volume Traded')
ford['Volume'].plot(label="Ford")
gm['Volume'].plot(label="General Motors")
plt.legend();
```

**Analyzing the volume traded plot we can see that Ford had a big spike in late 2013** \
**Analyze further by finding the exact date**

```python
ford['Volume'].idxmax()
```

```python
ford['Open'].plot(figsize = (12,6))
```

```python
#To analyze more fairly we need teh total market cap of each company which would be the total amount of money traded
```

```python
tesla['Total Traded'] = tesla['Open']*tesla['Volume']
ford['Total Traded'] = ford['Open']*ford['Volume']
gm['Total Traded'] = gm['Open']*gm['Volume']
```

```python
#Plot against time index
```

```python
tesla["Total Traded"].plot(label ="Tesla", figsize = (16,8))
ford["Total Traded"].plot(label ="Ford", figsize = (16,8))
gm["Total Traded"].plot(label ="General Motors", figsize = (16,8))

plt.legend();
```

```python
#We can see that there was a lot of money traded near early 2014, what happens to be that date

tesla['Total Traded'].idxmax()

#Put in another annotation image here
```

```python
#Plotting out moving averages
#Notice how the moving averages for 50 days and 200 days are cutoff since the first 50 and 200 days are ingnored since a mean
#cannot be calculated for them

gm['Moving Average of 50 days'] = gm['Open'].rolling(50).mean()
gm['Moving Average of 200 days'] = gm['Open'].rolling(200).mean()

gm[['Open', "Moving Average of 50 days","Moving Average of 200 days"]].plot(figsize = (16,10))
```

```python
#Lets now look for correlation and how these stocks may be related to each other using a scatter plot

from pandas.plotting import scatter_matrix
```

```python
car_company = pd.concat([tesla["Open"],gm["Open"],ford["Open"]], axis = 1)
```

```python
car_company.columns = ["Tesla Open","GM Open","Ford Open" ]
```

```python
scatter_matrix(car_company,figsize = (20,20), alpha = 0.2, hist_kwds = {'bins' : 20});

#From this we can see there may be some correlation between ford and gm but further research is needed in order to come
#to an appropriate conclusion
```
